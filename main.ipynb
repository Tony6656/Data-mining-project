2.2.3通过属性的相关关系来填补缺失值

利用线性回归方式，根据下载量和观看量拟合函数，对评分进行预测，将缺失值填充后观察到数据分布出现偏移。

```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
p = 'movies_dataset.csv'
data=pd.read_csv(p,dtype = 'str')
data['IMDb-rating'] = data['IMDb-rating'].astype(float)
k  = 0
for i in data['downloads']:
    if isinstance(i,float):
        k+=1
        continue
    else:
        if ',' in i:
            data['downloads'][k] = i.replace(',','')
        k+=1
k = 0
for i in data['views']:
    if isinstance(i,float):
        k+=1
        continue
    else:
        data['views'][k] = i.replace(',','')
        k+=1
data['downloads']=data['downloads'].astype(float)
data['views']=data['views'].astype(float)
res = []
for i in range(len(data['IMDb-rating'])):
    if np.isnan(data['IMDb-rating'][i]) or  np.isnan(data['downloads'][i]) or np.isnan(data['views'][i]):
        res.append(i)
plt.hist(data['IMDb-rating'],bins=40, facecolor="blue", edgecolor="black", alpha=0.7)
plt.xlabel('IMDb-rating')
plt.ylabel("frequency")
plt.show()
data_new1 = data.drop(index = data.index[res])

X = data_new1[['downloads','views']]
# X = data_new1['downloads']
y = data_new1['IMDb-rating']
#构建线性回归模型
model = LinearRegression()
model.fit(X, y)

for i in range(len(data['IMDb-rating'])):
    if np.isnan(data['IMDb-rating'][i]):
        x = [[data['downloads'][i],data['views'][i]]]
        pred = model.predict(x)
        data['IMDb-rating'][i] = round(pred[0], 1)


plt.hist(data['IMDb-rating'],bins=40, facecolor="blue", edgecolor="black", alpha=0.7)
plt.xlabel('IMDb-rating')
plt.ylabel("frequency")
plt.show()
```

2.2.4通过数据对象之间的相似性来填补缺失值

利用KNN方法衡量数据对象之间的相关性，相比于线性回归方法，填充前和填充后的数据分布没有发生较为明显的改变，符合预期。

```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier,KNeighborsRegressor
import numpy as np
p = 'movies_dataset.csv'
data=pd.read_csv(p,dtype = 'str')
data['IMDb-rating'] = data['IMDb-rating'].astype(float)
k  = 0
for i in data['downloads']:
    if isinstance(i,float):
        k+=1
        continue
    else:
        if ',' in i:
            data['downloads'][k] = i.replace(',','')
        k+=1
k = 0
for i in data['views']:
    if isinstance(i,float):
        k+=1
        continue
    else:
        data['views'][k] = i.replace(',','')
        k+=1
data['downloads']=data['downloads'].astype(float)
data['views']=data['views'].astype(float)
plt.hist(data['IMDb-rating'],bins=40, facecolor="blue", edgecolor="black", alpha=0.7)
plt.xlabel('IMDb-rating')
plt.ylabel("frequency")
plt.show()
def knn_missing_filled(x_train,y_train,test,k=5,dispersed=False):
    if dispersed:
        clf=KNeighborsClassifier(n_neighbors=k,weights="distance")
    else:
        clf=KNeighborsRegressor(n_neighbors=k,weights="distance")
    clf.fit(x_train,y_train)
    return test.index,clf.predict(test)

res = []
for i in range(len(data['IMDb-rating'])):
    if np.isnan(data['IMDb-rating'][i]) or  np.isnan(data['downloads'][i]) or np.isnan(data['views'][i]):
        res.append(i)

data_new1 = data.drop(index = data.index[res])

x = data_new1[['downloads','views']]
y = data_new1['IMDb-rating']

clf=KNeighborsRegressor(n_neighbors=5,weights="distance")
clf.fit(x,y)

for i in range(len(data['IMDb-rating'])):
    if np.isnan(data['IMDb-rating'][i]):
        x = np.array([[data['downloads'][i],data['views'][i]]])
        pred = clf.predict(x)
        data['IMDb-rating'][i] = round(pred[0], 1)

plt.hist(data['IMDb-rating'],bins=40, facecolor="blue", edgecolor="black", alpha=0.7)
plt.xlabel('IMDb-rating')
plt.ylabel("frequency")
plt.show()
```
